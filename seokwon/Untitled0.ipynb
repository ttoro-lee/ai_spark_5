```python
from sklearn.impute import KNNImputer

def impute_knn(df):
    imputer = KNNImputer(n_neighbors=5)
    df[['PM2.5']] = imputer.fit_transform(df[['PM2.5']])
    df[['기온(°C)']] = imputer.fit_transform(df[['기온(°C)']])
    df[['풍향(deg)']] = imputer.fit_transform(df[['풍향(deg)']])
    df[['풍속(m/s)']] = imputer.fit_transform(df[['풍속(m/s)']])
    df[['강수량(mm)']] = imputer.fit_transform(df[['강수량(mm)']])
    df[['습도(%)']] = imputer.fit_transform(df[['습도(%)']])
    return df
```

```python
X = []
y = []
for j in range(48, len(train)-72):
     X.append(train.iloc[j-48:j, 1:].values)
     y.append(train.iloc[j:j+72, 0].values)
X_train=np.array(X)
y_train = np.array(y)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout


batch_size = 32
epochs = 100
### 원래는 0.001    #### learning_rate 영향을 많이 받음
learning_rate = 0.0008

dropout_rate = 0.01

model = Sequential()
model.add(LSTM(units=16, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=32, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=64, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=128, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256, return_sequences=True))
model.add(Dropout(dropout_rate))
model.add(LSTM(units=256))
model.add(Dense(units=72, activation='linear'))


optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])

from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', min_delta=0.001, verbose=1, restore_best_weights=True, patience=5)
# 모델 학습
history = model.fit(X_train, y_train, epochs=epochs,callbacks=[es], batch_size=batch_size, validation_data=(X_val, y_val))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 모델 생성
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=72, activation='linear'))

# 모델 컴파일
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])

from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', min_delta=0, verbose=1, restore_best_weights=True, patience=5)

# 모델 학습
history = model.fit(X_train, y_train, epochs=50, batch_size=32, callbacks=[es], validation_data=(X_val, y_val))
```
