```python
from sklearn.impute import KNNImputer

def impute_knn(df):
    imputer = KNNImputer(n_neighbors=7)
    df[['PM2.5']] = imputer.fit_transform(df[['PM2.5']])
    df[['기온(°C)']] = imputer.fit_transform(df[['기온(°C)']])
    df[['풍향(deg)']] = imputer.fit_transform(df[['풍향(deg)']])
    df[['풍속(m/s)']] = imputer.fit_transform(df[['풍속(m/s)']])
    df[['강수량(mm)']] = imputer.fit_transform(df[['강수량(mm)']])
    df[['습도(%)']] = imputer.fit_transform(df[['습도(%)']])
    return df
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 모델 생성
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=72, activation='linear'))

# 모델 컴파일
model.compile(loss='mse', optimizer='adam', metrics=['mae'])

from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', min_delta=0, verbose=1, restore_best_weights=True, patience=5)

# 모델 학습
history = model.fit(X_train, y_train, epochs=50, batch_size=32, callbacks=[es], validation_data=(X_val, y_val))
```
