```python
def make_model(target):
    
    target = target

    X, Y = data_load(target)

    data = X.values
    data_y = Y.values

    timestep = 48 # 한 차원에 48시간

    x = np.array([data[i : i + timestep] for i in range( len(data) - timestep ) ])
    y = np.array([data_y[i + timestep] for i in range( len(data_y) - timestep ) ])

    x_train, x_val, x_test = x[:len(x)-216], x[len(x)-216: len(x)-72], x[len(x)-72: len(x)]
    y_train, y_val, y_test = y[:len(y)-216], y[len(y)-216: len(y)-72], y[len(y)-72: len(y)]

    keras.backend.clear_session()

    il = keras.layers.Input(shape=(x_train.shape[1], x_train.shape[2]))

    hl = keras.layers.LSTM(16, return_sequences=True)(il)
    hl = keras.layers.GRU(32, return_sequences=True)(hl)
    hl = keras.layers.Flatten()(hl)
    hl = keras.layers.Dense(128, activation='swish')(hl)
    ol = keras.layers.Dense(1)(hl)

    model = keras.models.Model(inputs=il, outputs=ol)

    model.compile(loss=keras.losses.mae, optimizer='adam')

    es = keras.callbacks.EarlyStopping(monitor='val_loss',
                                      min_delta=0,
                                      verbose=1,
                                      restore_best_weights=True,
                                      patience=5)

    mc = keras.callbacks.ModelCheckpoint(monitor='val_loss',
                        filepath=f'./model/{target}_model.h5',
                        save_best_only=True,
                         verbose=1)

    hist = model.fit(x_train, y_train, validation_data=(x_val, y_val), callbacks=[es, mc], verbose=1, epochs=100)
    
    return
```
